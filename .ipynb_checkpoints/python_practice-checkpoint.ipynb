```python
s1 = "Abhijeet ugale"

# Positive Indexing
print("Positive Indexing:")
print(s1[0])   # A
print(s1[1])   # b
print(s1[2])   # h
print(s1[3])   # i
print(s1[4])   # j
print(s1[5])   # e
print(s1[6])   # e
print(s1[7])   # t
print(s1[8])   # (space)
print(s1[9])   # u
print(s1[10])  # g
print(s1[11])  # a
print(s1[12])  # l
print(s1[13])  # e

# Negative Indexing
print("\nNegative Indexing:")
print(s1[-1])   # e
print(s1[-2])   # l
print(s1[-3])   # a
print(s1[-4])   # g
print(s1[-5])   # u
print(s1[-6])   # (space)
print(s1[-7])   # t
print(s1[-8])   # e
print(s1[-9])   # e
print(s1[-10])  # j
print(s1[-11])  # i
print(s1[-12])  # h
print(s1[-13])  # b
print(s1[-14])  # A
```
